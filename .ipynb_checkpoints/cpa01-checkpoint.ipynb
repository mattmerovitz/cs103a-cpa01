```python
# This dataset is pupillometry data from the lab that I work for at Brandeis (Memory and Cognition Lab). 
# As the participant goes through the study, a pupillometer collects the left pupil size of the participant for analysis.
# The dataset can be found at https://drive.google.com/file/d/1nb81iat_qbOPsL4T6G0GCIfWyaGfvNdQ/view?usp=sharing.
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('tcv001_mod_cpa01.txt', sep='\t', lineterminator='\r', low_memory=False)
```

```python
df.describe()
```

```python
print(df.columns)
print(df.index)
df['LEFT_PUPIL_SIZE'] = pd.to_numeric(df['LEFT_PUPIL_SIZE'], errors='coerce')
df['LEFT_PUPIL_SIZE'].plot()
pvt = pd.pivot_table(data=df,columns=['stimname'],index=['SAMPLE_INDEX'],values=['LEFT_PUPIL_SIZE'])
pl = pvt.plot()
pl.legend(bbox_to_anchor=(1.0, 1.0))
gb = df.groupby(['stimname'])
```

```python
#The top plot shows all of the subject's left pupil sizes for the entire length of the study. 
#The bottom plot shows the same data but over individual passages in the study. 
#The goal of this is to see the difference in pupil size as different passages are read to the participant.
```
